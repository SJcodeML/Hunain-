Core Data Type Details


Numeric Types:
Integer/int: Whole numbers of unlimited precision.
float: Real numbers with decimal points, typically accurate to 15–17 decimal places.
complex: Numbers with a real and an imaginary part (e.g., 3 + 4j) 3 is real number , 4 is imaginary part 

# ✅ int (Integer)
# Whole numbers, positive or negative, without decimals
age = 25
population = 7_900_000_000  # underscores for readability
temperature_change = -12
print(f"int examples: {age}, {population}, {temperature_change}")
print(f"Type: {type(age)}\n")

# A float is any number with a decimal point
price = 12.99      # price of an item
temperature = -3.5 # freezing-point temperature in °C
average_score = 87.5

print("Examples:")
print(f"Price: ${price}")           # → $12.99
print(f"Temp: {temperature}°C")  # → -3.5°C
print(f"Average: {average_score}%") # → 87.5%

print(f"\nAll are floats: {type(price)}, {type(temperature)}, {type(average_score)}")

# ✅ complex (Complex number)
# Represented as a + bj, where 'a' is real, 'b' is imaginary, and 'j' is √(-1)
z1 = 3 + 4j
z2 = -2.5 + 0.7j
z3 = 5j  # purely imaginary
print(f"complex examples: {z1}, {z2}, {z3}")
print(f"Real part of {z1}: {z1.real}, Imaginary part: {z1.imag}")
print(f"Type: {type(z1)}")

Sequence Types:
str: Immutable sequences of Unicode characters.
list: Ordered and mutable collections of items (can be changed after creation).
tuple: Ordered and immutable collections (cannot be changed).
range: Represents an immutable sequence of numbers, often used in loops

Mapping Type (dict):
Stores data in key-value pairs. Keys must be unique and immutable

Set Types:
set: Unordered collection of unique items; mutable.
frozenset: Immutable version of a set.

Essential Functions
type(variable): Returns the data type of the specified object.
isinstance(object, type): Checks if an object is an instance of a specific class or type.
Type Casting: You can convert between types using constructors like int(), float(), str(), list(), and tuple()
