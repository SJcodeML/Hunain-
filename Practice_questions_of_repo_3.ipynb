1. Swapping Variables
Write a Python program that takes two variables, a and b, assigns them any numeric values, and then swaps their values.

Challenge: Try to do this both with a third temporary variable and without using a third variable (the Pythonic way).

```python
a = 5
b = 10

# Using a third variable
temp = a
a = b
b = temp

print(f"After swap: a = {a}, b = {b}")
```

Without a temporary variable (The Pythonic Way):

```python
a = 5
b = 10

# Tuple unpacking assignment
a, b = b, a

print(f"After swap: a = {a}, b = {b}")
```

2. Calculating the Area of a CircleCreate a program where you declare a variable radius and assign it a decimal value. Use the formula $\text{Area} = \pi r^2$ to calculate the area and print a user-friendly message. (You can approximate $\pi$ as 3.14159).

```python
radius = 4.5
pi = 3.14159

# Calculate area
area = pi * (radius ** 2)

print(f"The area of a circle with radius {radius} is {area}")
```

3. Gathering User Bio
Write a script that asks a user for their first name, last name, and birth year using the input() function. Calculate their current age dynamically based on the current year, and print a single formatted string that says:
"Hello [First Name] [Last Name], you are [Age] years old!"

```python
# Gathering inputs
first_name = input("Enter your first name: ")
last_name = input("Enter your last name: ")
birth_year = int(input("Enter your birth year: "))

# Dynamic age calculation (Assuming current year is 2026)
current_year = 2026
age = current_year - birth_year

# Formatted output
print(f"Hello {first_name} {last_name}, you are {age} years old!")
```

4. Playing with Remainder and Division
Given a total number of minutes (e.g., 135), write a script that calculates and prints how many full hours and remaining minutes that equals.

Hint: Use integer division (//) and the modulo operator (%).

```python
total_minutes = 135

# Floor division gets the whole hours
hours = total_minutes // 60

# Modulo gets the remaining minutes
minutes = total_minutes % 60

print(f"{total_minutes} minutes is equal to {hours} hour(s) and {minutes} minute(s).")
```

1. Re-assigning with In-Place Operators
Start with a tracking variable score = 100. Write a sequence of operations using augmented assignment operators (+=, -=, *=, /=, =) to complete the following story:

The player gains 50 points.

The player hits a multiplier that doubles their score.

The player takes a penalty that reduces their score by 15%.

Print the final score as an integer.

```python
score = 100

score += 50   # 100 + 50 = 150
score *= 2    # 150 * 2 = 300
score *= 0.85 # Reduces score by 15% (keeps 85%) -> 300 * 0.85 = 255.0

# Converting to integer for clean output
print("Final Score:", int(score))
```

2. Identity vs. Equality Testing
Explain the output of the following code block. Why do print(x == y) and print(x is y) give different results?

```python
x = [1, 2, 3]
y = [1, 2, 3]

print(x == y)
print(x is y)

z = x
print(x is z)
```

Explanation:

x == y is True because the values inside both lists are identical.

x is y is False because x and y point to two different memory locations. They are distinct objects in memory.

x is z is True because assigning z = x creates a reference copy. Both x and z now point to the exact same object in memory.

3. Logical Operator Precedence
Predict the boolean outcome (True or False) of the following expression without running it in an IDE, keeping the rules of operator precedence (not, then and, then or) in mind. Once you have your answer, write code to verify it.

```python
a = True
b = False
c = True

result = a or not b and c and not a
print(result)
```

```python
# Original expression: True or not False and True and not True
# 1. Evaluate 'not' operators first:
# not False becomes True
# not True becomes False
# Expression is now: True or True and True and False

# 2. Evaluate 'and' operators next (left to right):
# True and True becomes True -> Expression: True or True and False
# True and False becomes False -> Expression: True or False

# 3. Evaluate 'or' operator last:
# True or False becomes True

# Verification code:
a = True
b = False
c = True

result = a or not b and c and not a
print("Result:", result) # Output: Result: True
```
