# **1-Numeric Types:**
a- Integer/int: Whole numbers of unlimited precision.


b- float: Real numbers with decimal points, typically accurate to 15–17 decimal places.



c- complex: Numbers with a real and an imaginary part (e.g., 3 + 4j) 3 is real number , 4 is imaginary part

✅ int (Integer)
Arbitrary-precision whole numbers (no fractional part, no size limit):

```python
a = 42              # positive integer
b = -17             # negative integer
c = 0               # zero
d = 123_456_789     # underscores for readability (same as 123456789)
e = 0b1010          # binary literal → 10
f = 0xFF            # hexadecimal literal → 255
g = 10**100         # a googol — perfectly valid int!
```

💡 Note: Unlike many languages, Python int has no overflow — it scales to fit any size.

✅ float (Floating-point)
IEEE 754 double-precision real numbers (≈15–17 decimal digits of precision):

```python
x = 3.14159         # decimal float
y = -0.001          # negative small float
z = 2.5e3           # scientific notation → 2500.0
w = 1.2e-4          # → 0.00012
t = 0.1 + 0.2       # → 0.30000000000000004 (due to binary representation!)
```

✅ complex (Complex number)
Numbers with real and imaginary parts (a + bj, where j is √−1):

```python
c1 = 3 + 4j         # real=3, imag=4
c2 = -2.5 + 0j      # real only → still type complex
c3 = 7j             # purely imaginary (equivalent to 0 + 7j)
c4 = complex(1, 2)  # constructor → 1+2j
c5 = 5 - 1.5j       # mixed int/float components

# Access parts:
print(c1.real)   # 3.0
print(c1.imag)   # 4.0
print(abs(c1))   # magnitude = √(3² + 4²) = 5.0
```

🧠 Fun fact: j (not i) is used in Python because i is often used as a loop variable — j avoids ambiguity.

```python
# ✅ int (Integer)
# Whole numbers, positive or negative, without decimals
age = 25
population = 7_900_000_000  # underscores for readability
temperature_change = -12
print(f"int examples: {age}, {population}, {temperature_change}")
print(f"Type: {type(age)}\n")

# A float is any number with a decimal point
price = 12.99      # price of an item
temperature = -3.5 # freezing-point temperature in °C
average_score = 87.5

print("Examples:")
print(f"Price: ${price}")           # → $12.99
print(f"Temp: {temperature}°C")  # → -3.5°C
print(f"Average: {average_score}%") # → 87.5%

print(f"\nAll are floats: {type(price)}, {type(temperature)}, {type(average_score)}")

# ✅ complex (Complex number)
# Represented as a + bj, where 'a' is real, 'b' is imaginary, and 'j' is √(-1)
z1 = 3 + 4j
z2 = -2.5 + 0.7j
z3 = 5j  # purely imaginary
print(f"complex examples: {z1}, {z2}, {z3}")
print(f"Real part of {z1}: {z1.real}, Imaginary part: {z1.imag}")
print(f"Type: {type(z1)}")
```

**d- Range(range)**: Immutable sequence of numbers (lazy-evaluated)
Represents a sequence of numbers.

```python
r = range(2, 10, 3)  # start=2, stop=10, step=3 → values: 2, 5, 8
# r[0] = 100         # ❌ TypeError: 'range' object does not support item assignment

print(list(r))       # [2, 5, 8] — converted to list to see values
for i in r:
    print(i)         # prints 2, then 5, then 8

# Often used in loops:
for i in range(3):
    print(f"Loop {i}")  # Loop 0, Loop 1, Loop 2
```

✅ Immutable & memory-efficient: Generates numbers on demand, not stored in memory all at once.

# **a- Set(set)**

🔑**Key ideas upfront:**

Set Types:

set: Unordered collection of unique items; mutable.

frozenset: Immutable version of a set.

set: mutable, unordered, no duplicates, elements must be hashable (i.e., immutable).

frozenset: immutable version of set — can be used as dictionary keys or in other sets.

Neither supports indexing (s[0]) or slicing (s[1:3]) — they’re not sequences.

Both support mathematical set operations: union (|), intersection (&), difference (-), symmetric difference (^).

```python
# Creating a set (order is not preserved; duplicates are ignored)
s = {3, 1, 4, 1, 5}  # → {1, 3, 4, 5}
print(s)  # e.g., {1, 3, 4, 5} — order may vary

# Also created from iterables
s2 = set("hello")      # → {'h', 'e', 'l', 'o'} (only one 'l')
s3 = set([1, 2, 2, 3]) # → {1, 2, 3}

# Mutation allowed ✅
s.add(9)
s.remove(3)
s.discard(99)  # no error if missing
s.update([10, 11])  # adds multiple

print(s)  # e.g., {1, 4, 5, 10, 11}
```

❌ Invalid: s[0] → TypeError: 'set' object is not subscriptable

❌ Invalid: {[1, 2], [3, 4]} → TypeError: unhashable type: 'list'

# b-FrozenSet

```python
# Immutable — created like set(), but with 'frozenset()'
fs = frozenset([1, 2, 2, 3])   # → frozenset({1, 2, 3})
print(fs)  # always prints sorted-like, but it's still unordered internally

# Can be used where immutability matters ✅
d = {fs: "my frozen set"}      # works — dict key!
other_set = {1, 2, fs}         # works — element of another set!

# But mutation fails ❌
# fs.add(4)        → AttributeError: 'frozenset' object has no attribute 'add'
# fs.remove(1)     → AttributeError
```

🔍 Bonus: hash(fs) works → hash(frozenset({1,2})) returns an int.
But hash({1,2}) → TypeError: unhashable type: 'set'.

# **b- Mapping Type (dict):**
Stores data in key-value pairs. Keys must be unique and immutable

🔑 What is a dict?
A dict is a mutable, unordered (as of Python 3.7+, insertion-ordered by default, but order is still considered an implementation detail for logic — don’t rely on it for algorithmic ordering) collection of key-value pairs, where:




✅ Keys must be immutable (e.g., str, int, float, bool, tuple — if all its
elements are immutable).

❌ Mutable types like list, set, or another dict cannot be keys, because they can change — breaking hash consistency.

✅ Values can be anything: mutable or immutable, duplicates allowed.

✅ Keys are unique: assigning to an existing key overwrites the value.

Internally, dict uses a hash table → average O(1) lookup/insert/delete.

🧪 Examples

**1. Basic creation & access**


```python
person = {"name": "Alice", "age": 30, "city": "Tokyo"}
print(person["name"])        # ✅ 'Alice'
print(person.get("email"))   # ✅ None (safe access)
```

2. Immutable keys — what works vs. what fails

```python
# ✅ Valid keys
valid_dict = {
    "hello": 1,
    42: "answer",
    (1, 2): "tuple key",
    True: "boolean key"
}

# ❌ Invalid — raises TypeError: unhashable type
# invalid_dict = {[1, 2]: "list key"}      # TypeError!
# invalid_dict = {{'a': 1}: "dict key"}    # TypeError!
# invalid_dict = {{1, 2}: "set key"}       # TypeError!
```

3. Mutability in action (dict itself is mutable)

```python
person["country"] = "Japan"     # ✅ Add new key-value
person["age"] = 31              # ✅ Update existing value
del person["city"]              # ✅ Remove key
person.clear()                  # ✅ Empty the dict
```

4. Keys must be unique — duplicates are silently overwritten

```python
d = {"x": 1, "x": 2}
print(d)  # {'x': 2} — second assignment wins
```

5. Real-world usage pattern: counting with dict

```python
words = ["apple", "banana", "apple", "cherry", "banana", "apple"]
count = {}
for word in words:
    count[word] = count.get(word, 0) + 1
print(count)  # {'apple': 3, 'banana': 2, 'cherry': 1}
```


💡** Bonus Insight: Why immutability matters for keys**

Python computes a hash of each key to place it in the hash table. If a key could change after insertion (e.g., a list), its hash would change → Python wouldn’t know where to find it. That’s why only immutable objects — whose hash stays constant — are allowed as keys.

You can check if something is hashable:



```python
print(hash("hello"))    # ✅ works → returns int
print(hash([1, 2]))     # ❌ TypeError: unhashable type
```

# **Type Casting**

Type casting is the process of converting a value of one data type to another data type. Python supports several types of type casting, including:

Implicit Type Casting: Python automatically converts a value of one data type to another data type when necessary. For example, when you add an integer and a float, Python converts the integer to a float.

Explicit Type Casting: You can use functions like int(), float(), str(), and bool() to explicitly convert a value of one data type to another data type.

```python
i: int = 10
print("Value of i = " + str(i) + ",     Type of i = " + str(type(i)))

j: float = 20.6

f: float = i + j #Implicit Type Casting
print("Value of f = " + str(f) + ",   Type of f = " + str(type(f)))

f1: float = 66.89
print("Value of f1 = " + str(f1) + ", Type of i = " + str(type(f1)))

i1: int = int(f1) #When ever you type cast a float value into an integer it truncate
             #the decimal part and only keeps the whole number
print("Value of i1 = " + str(i1) + ",    Type of i = " + str(type(i1)))

s: str = "25.8"
f2: float = float(s)
print("Value of f2 = " + str(f2) + ",  Type of i = " + str(type(f2)))

#uncomment the below line of code to see error
#i2 = int(s) #correct this error by casting with float()
#print("Value of i2 = " + str(i2) + ", Type of i = " + str(type(i2)))

```
